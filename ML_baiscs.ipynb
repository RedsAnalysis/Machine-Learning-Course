```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("online_shoppers_intention.csv")
data
```

```python
data["Revenue"].unique()
```

```python
data["Revenue"] = (data["Revenue"]).astype(int)
data["Weekend"] = (data["Weekend"]).astype(int)
```

```python
data
```

```python
data["Weekend"].unique()
```

```python
# Apply one-hot encoding to 'VisitorType'
visitor_dummies = pd.get_dummies(data['VisitorType'], prefix='VisitorType')

# View the new DataFrame with dummy variables
print(visitor_dummies.head())
```

```python
# Concatenate the original DataFrame and the new dummy variable DataFrame
data = pd.concat([data, visitor_dummies], axis=1)

# Optionally, you can drop the original 'VisitorType' column if you no longer need it
# df = df.drop('VisitorType', axis=1)
```

```python
data = data.drop('VisitorType', axis=1)
```

```python
data
```

```python
data["VisitorType_New_Visitor"] = (data["VisitorType_New_Visitor"]).astype(int)
data["VisitorType_Other"] = (data["VisitorType_Other"]).astype(int)
data["VisitorType_Returning_Visitor"] = (data["VisitorType_Returning_Visitor"]).astype(int)
```

```python
data["Month"].unique()
```

```python
month_to_int = {
    'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4,
    'May': 5, 'June': 6, 'Jul': 7, 'Aug': 8,
    'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12}
```

```python
data['Month_Ordinal'] = data['Month'].map(month_to_int)
```

```python
import numpy as np

# Cyclic encoding
data['Month_Cos'] = np.cos(2 * np.pi * data['Month_Ordinal']/12)
data['Month_Sin'] = np.sin(2 * np.pi * data['Month_Ordinal']/12)
```

```python
data
```

```python
data = data.drop('Month', axis=1)
```

```python
for label in data.columns[:-1]:
  plt.hist(data[data["Revenue"]==1][label], color = "green", label = "Bought", alpha = 0.7, density = True)
  plt.hist(data[data["Revenue"]==0][label], color = "red", label = "Didn't buy", alpha = 0.7, density = True)
  plt.title(label)
  plt.ylabel("Probability")
  plt.xlabel(label)
  plt.legend()
  plt.show()
```

```python
df = pd.read_csv("qsar_aquatic_toxicity.csv", names = cols, delimiter = ";")
df
```

```python
train, val , test = np.split(df.sample(frac=1), [int(0.6*len(df)), int(0.8*len(df))])
```

```python
import copy
```

```python
def get_xy (DataFrame, x_label=None, y_label=None):
  DataFrame = copy.deepcopy(DataFrame)
  if x_label is None:
    X = DataFrame[[i for i in DataFrame.columns if i != y_label]].values
  else:
      if len(x_label) == 1:
        X = DataFrame[x_label[0]].values.reshape(-1.1)
      else:
        X = DataFrame[x_label].values

  y = DataFrame[y_label].values.reshape(-1,1)
  data = np.hstack((X,y))

  return data, X, y
```

```python
full_data, X_train, y_train = get_xy(train, x_label = df.columns[:-1], y_label= ["LC50"])
full_data, X_val, y_val = get_xy(val, x_label = df.columns[:-1], y_label= ["LC50"])
full_data, X_test, y_test = get_xy(test, x_label = df.columns[:-1], y_label= ["LC50"])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_model = LinearRegression()
linear_model.fit(X_train,y_train)
```

```python
linear_model.score(X_test, y_test)
```
